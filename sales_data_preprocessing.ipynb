```python
import pandas as pd
raw_path=r"C:\Users\guntu\OneDrive\Desktop\sales-performance-project\data\raw\Sample_ Superstore.csv"
df = pd.read_csv(raw_path)

# Clean column names
df.columns = df.columns.str.strip().str.lower().str.replace(" ", "_")

# Robust date parsing (handles / and - automatically)
df['order_date'] = pd.to_datetime(
    df['order_date'].astype(str).str.strip(),
    errors='coerce',
    infer_datetime_format=True
)

df['ship_date'] = pd.to_datetime(
    df['ship_date'].astype(str).str.strip(),
    errors='coerce',
    infer_datetime_format=True
)

# Remove rows where dates are invalid
df = df.dropna(subset=['order_date', 'ship_date'])

# Create time features
df['order_year'] = df['order_date'].dt.year
df['order_month'] = df['order_date'].dt.month
df['order_month_name'] = df['order_date'].dt.month_name()

# Profit Margin KPI
df['profit_margin'] = (df['profit'] / df['sales']) * 100


print("Cleaning completed. Rows:", len(df))
# Save safely with proper quoting
df.to_csv(
    "superstore_cleaned.csv",
    index=False,
    quoting=1   # QUOTE_ALL
)
```

```python
import pandas as pd
raw_path=r"C:\Users\guntu\OneDrive\Desktop\sales-performance-project\data\raw\Sample_ Superstore.csv"
df = pd.read_csv(raw_path)


def fix_date(col):
    col = col.astype(str).str.strip()
    col = col.str.replace('/', '-', regex=False)
    return pd.to_datetime(col, errors='coerce', format='mixed')

# 3. Convert date columns safely
df['Order Date'] = fix_date(df['Order Date'])
df['Ship Date']  = fix_date(df['Ship Date'])

# 4. Check missing values (optional debug)
print("Missing dates:")
print(df[['Order Date','Ship Date']].isna().sum())

# 5. Drop only rows where BOTH dates are missing
df = df[~(df['Order Date'].isna() & df['Ship Date'].isna())]

# 6. Create business columns
df['order_year'] = df['Order Date'].dt.year
df['order_month'] = df['Order Date'].dt.month
df['order_month_name'] = df['Order Date'].dt.month_name()
df['profit_margin'] = (df['Profit'] / df['Sales']) * 100

# 7. Final check
print("Final row count:", len(df))


```
